## PCA iris Example

```python
X = iris.data # four columns of data for 150 rows
y = iris.target # one column of label for 150 rows
pca = PCA(n_components=0.95) # 95% of total variace.
pca.fit(X)
X_reduced = pca.transform(X) # similar to predict but works on the data itself
print(X_reduced.shape) # four columns are reduced to two. 150*4 --> 150*2
print(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=4)
pca.fit(X)
plt.scatter(range(pca.n_components_), pca.explained_variance_ratio_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn1 = KNeighborsClassifier(n_neighbors=5)
knn.fit(X,y) #fit on original data set
pca = PCA(n_components=1)
pca.fit(X)
X_reduced = pca.transform(X)
knn1.fit(X_reduced, y)
```

```python
from sklearn.metrics import accuracy_score
print('Original 4 features', accuracy_score(knn.predict(X),y))
print('reduced PCA features', accuracy_score(knn1.predict(X_reduced),y))
```

# Pipelines

```python
from sklearn.pipeline import Pipeline
```

```python
pca = PCA()
knn = KNeighborsClassifier()
```

```python
myPipe = Pipeline(steps=[('pca', pca), ('knn', knn)])
```

```python
myPipe
```

# Grid Search

```python
param_grid = {
    'pca__n_components': [4,3,2,1],
    'knn__n_neighbors': [3,5,7],
    }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
search = GridSearchCV(myPipe, param_grid, n_jobs=1)
search.fit(X,y) # in real world, this step can run for hours
```

```python
search.best_params_ # the output explains that if we use 5 neighbors or 4 components that we get best performance/outut
```
