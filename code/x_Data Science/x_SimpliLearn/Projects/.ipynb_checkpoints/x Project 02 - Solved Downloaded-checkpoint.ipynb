```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('S:\\SandeepG\\Official\\DataScience\\SimpliLearn\\Data Science with Python\\Projects\\Project 02 - Solved Downloaded\\ml-1m\\ml-1m\\ratings.dat', sep='::', names=['user_id','item_id','rating','timestamp'])
df.head()
```

```python
movie_titles = pd.read_csv('S:\\SandeepG\\Official\\DataScience\\SimpliLearn\\Data Science with Python\\Projects\\Project 02 - Solved Downloaded\\ml-1m\\ml-1m\\movies.dat', sep='::', names=['item_id','title','genre'])
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python
df.describe()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['number_of_ratings'] = df.groupby('title')['rating'].count()
ratings.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
ratings['rating'].hist(bins=50)
```

```python
ratings['number_of_ratings'].hist(bins=60)
```

```python
import seaborn as sns
sns.jointplot(x='rating', y='number_of_ratings', data=ratings)
```

```python
df.columns
movie_matrix = df.pivot_table(index='user_id', columns='title', values='rating')
#df['title'].unique()
movie_matrix.head(10)
```

```python
ratings.sort_values('number_of_ratings', ascending=False).head(10)
```

```python
AFO_user_rating = movie_matrix['Air Force One (1997)']
contact_user_rating = movie_matrix['Contact (1997)']
```

```python
AFO_user_rating.head()
contact_user_rating.head()
```

```python
similar_to_air_force_one=movie_matrix.corrwith(AFO_user_rating)
```

```python
similar_to_air_force_one.head()
```

```python
similar_to_contact = movie_matrix.corrwith(contact_user_rating)
```

```python
similar_to_contact.head()
```

```python
corr_contact = pd.DataFrame(similar_to_contact, columns=['Correlation'])
corr_contact.dropna(inplace=True)
print(corr_contact.head())

corr_AFO = pd.DataFrame(similar_to_air_force_one, columns=['correlation'])
corr_AFO.dropna(inplace=True)
print(corr_AFO.head())
```

```python
corr_AFO = corr_AFO.join(ratings['number_of_ratings'])
corr_contact = corr_contact.join(ratings['number_of_ratings'])
corr_AFO .head()
corr_contact.head()
```

```python
corr_AFO[corr_AFO['number_of_ratings'] > 100].sort_values(by='correlation', ascending=False).head(10)
```

```python
corr_contact[corr_contact['number_of_ratings'] > 100].sort_values(by='Correlation', ascending=False).head(10)
```
