## <font color=blue>Exploratory Data Analysis

#### <font color=blue> Task 01 : Which movies have maximum views/ratings?

```python
# importing the dataset
import pandas as pd
ratingsDF = pd.read_csv("S:\\SandeepG\\Official\\DataScience\\SimpliLearn\\Data Science with Python\\Projects\\Project 03 - Amazon\\Amazon - Movies and TV Ratings.csv",  sep=',', engine='python', header='infer')
ratingsDF
```

```python
ratingsDF.describe()
```

```python
ratingsDF_described = ratingsDF.describe()
```

```python
ratingsDF_transposed = ratingsDF_described.transpose()
```

```python
ratingsDF_transposed.reset_index(level=0, inplace=True)
ratingsDF_transposed = ratingsDF_transposed.rename({"index":"movie_name"}, axis='columns') 
```

```python
ratingsDF_transposed[['movie_name', 'count', 'mean']].sort_values(by='count', ascending=False).head(1)
```

#### <font color=green> Result 01 : Movie 127 has maximum views (2313) with mean rating of 4.111976

#### <font color=blue> Task 02 : What is the average rating for each movie? Define the top 5 movies with the maximum ratings.

```python
ratingsDF_transposed[['movie_name', 'count', 'mean']].sort_values(by='count', ascending=False)
```

#### <font color=green> Result 02 : Movie127, Movie140, Movie16, Movie103, Movie29 has maximum ratings.

#### <font color=blue> Task 03 : Define the top 5 movies with the least audience.

```python
ratingsDF_transposed[['movie_name', 'count', 'mean']].sort_values(by='count', ascending=True)
```

#### <font color=green> Result 03 : Movie1, Movie71, Movie145, Movie69, Movie68 has least audience.
