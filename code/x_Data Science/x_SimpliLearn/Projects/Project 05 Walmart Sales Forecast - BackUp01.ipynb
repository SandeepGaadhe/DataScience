#### <font color=blue> Question : Which store has maximum standard deviation i.e., the sales vary a lot. Also, find out the coefficient of mean to standard deviation?

```python
import numpy as np

print(df_rawdata.groupby("Store") \
      .apply(np.std)[['Weekly_Sales_InMillions']]\
      .sort_values(by = 'Weekly_Sales_InMillions', ascending = False)\
      .head())
```

#### <font color=green>Answer : Store 14 has maximum std deviation

```python
# calculating mean and std
df_mean_std = df_rawdata.groupby(['Store'], as_index=False).agg({'Weekly_Sales_InMillions':['mean','std']})

# calculating coefficient of mean to std dev
df_mean_std['Weekly_Sales_InMillions', 'Cv'] = df_mean_std['Weekly_Sales_InMillions', 'std']/abs(df_mean_std['Weekly_Sales_InMillions', 'mean'])

# sort desc by std_dev
df_mean_std.sort_values([('Weekly_Sales_InMillions', 'std')], ascending=False).head(1)
```

#### <font color=green>Answer : Above displays store with max std_dev and its coefficient of mean to std_dev (Cv)
