```python
import pandas as pd
abstract = list()
for pnum in pd.read_csv('mspats.csv')['PatNum'].values:
    url = 'https://patents.google.com/patent/US' + ''.join(pnum.split(','))
    page = requests.get(url)
    soup = BeautifulSoup(page.contect,'html.parser')
    page.close()
    abstract.append(list([tag for tag in soup.findAll('abstract')][0].stripping_strings)[0])
```
