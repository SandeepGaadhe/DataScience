### <font color=blue>Applying KNN model for item-based (movie-based) recommendation

```python
import os
import time

# data science imports
import math
import numpy as np
import pandas as pd
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

# utils import
from fuzzywuzzy import fuzz

# visualization imports
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# viewing raw data once again 
ratingsDF
```

```python
# unpivoting the dataset so that all user and all movies with respective rating can be converted into rows for easier
# extraction and analysis
ratingsUnpivot = ratingsDF.melt(id_vars=['user_id'], var_name='movie_name', value_name='movie_rating')
ratingsUnpivot
```

```python
# getting count of users and movies that have been rated at least once
# this is to identify the subset where neither the user nor the movie has been rated at all and cannot
# be included in recommendation system dataset
unique_users = len(ratingsUnpivot.user_id.unique())
unique_items = len(ratingsUnpivot.movie_name.unique())
expected_rating_count = unique_users * unique_items
print('expected_rating_count : ', expected_rating_count) # 998688 i.e. if every user rated every movie than total ratings received would be 998688
received_rating_count = ratingsUnpivot['movie_rating'].count()
print('received_rating_count : ', received_rating_count) # out of 998688 only 5000 is the count of rating received across movies and users
print('% of rating received : ', (received_rating_count/expected_rating_count)*100)
```

```python
# let's group by movie_rating 
movie_rating_cnt = pd.DataFrame(ratingsUnpivot.groupby('movie_rating').size(), columns=['count'])
movie_rating_cnt
```

```python
# get the count across movies and users which do not have any rating
not_rated_cnt = expected_rating_count - received_rating_count

# append the count of movies which are not rated to dataframe movie_rating_cnt
movie_rating_cnt = movie_rating_cnt.append(
    pd.DataFrame({'count': not_rated_cnt}, index=[0.0]),
    verify_integrity=True,
).sort_index()
movie_rating_cnt
```

```python
# add logarithmic count
movie_rating_cnt['log_count'] = np.log(movie_rating_cnt['count'])
movie_rating_cnt
```

```python
# visualize the rating over logarithmic count
ax = movie_rating_cnt[['count']].reset_index().rename(columns={'index': 'rating score'}).plot(
    x='rating score',
    y='count',
    kind='bar',
    figsize=(12, 8),
    title='Logarithmic Count of rating',
    logy=True,
    fontsize=12,
)
ax.set_xlabel("movie ratings")
ax.set_ylabel("rating frequency")
```

```python
# plot rating frequency of all movies in logarithmic scale
ax = movie_ratingFreq \
    .sort_values('rating_freq', ascending=False) \
    .reset_index(drop=True) \
    .plot(
        figsize=(12, 8),
        title='Plot for rating frequency of all movies in logarithmic scale',
        fontsize=12,
        logy=True
    )
ax.set_xlabel("movie_name")
ax.set_ylabel("rating_freq (in Log Scale)")
```

```python
# identify the outliers by knowing bare minimum threshold for rating_freq
movie_ratingFreq['rating_freq'].quantile(np.arange(1, 0.25, -0.05))
```

```python
# based on above quatile output, it is clear that we will have to keep threshold as 2 to cover 50% of our data 
# i.e. considering only those movies that has been rating more than once in our recommedation system

rating_freq_threshold = 2
movies_for_recSystem = movie_ratingFreq.query('rating_freq >= @rating_freq_threshold')

print('total movies in original data set: ', movie_ratingFreq.shape[0])
print('movies considered for recommendation system : ', movies_for_recSystem.shape[0])
```

```python
# identifying the movies to be considered for our recommendation system
movies_for_recSystem
```

```python
# identifyin the active user that rated movie(s)

ratingsUnpivot_for_recSystem = ratingsUnpivot[ratingsUnpivot.movie_name.isin(movies_for_recSystem['movie_name'].unique())]

df_users_rating_cnt = pd.DataFrame(ratingsUnpivot_for_recSystem.groupby('user_id').count())[['movie_rating']]

df_users_rating_cnt = df_users_rating_cnt.rename(\
                                                   {"movie_rating" : "no_of_movies_rated"
                                                   }, \
                                                   axis='columns') 

print(df_users_rating_cnt.sort_values(by = 'no_of_movies_rated', ascending = False))
```

```python
# understanding the distribution of active users
df_users_rating_cnt['no_of_movies_rated'].quantile(np.arange(1, 0.9, -0.005))
```

```python
# since most of the user has rated either 1 or 0 movies, we will have to consider only those users as active
# which has rating at least 02 movies
no_of_movies_rated_threshold = 2
users_for_recSystem = df_users_rating_cnt.query('no_of_movies_rated >= @no_of_movies_rated_threshold')

print('total users in original data set: ', ratingsDF.shape[0])
print('users considered for recommendation system : ', users_for_recSystem.shape[0])
```

```python
# plotting the active users and movies under consideration for rec system over a matrix
rec_movie_user_matrix = ratingsUnpivot_for_recSystem.pivot(index='movie_name', columns='user_id', values='movie_rating').fillna(0)

# create mapper from movie title to index
movie_to_idx = {
    movie: i for i, movie in 
    enumerate(movies_for_recSystem['movie_name'].unique())
}

# transform matrix to scipy sparse matrix
movie_user_mat_sparse = csr_matrix(rec_movie_user_matrix.values)
```

```python
# we will apply KNN model and Cosine metric to understand the distance difference between two movies. Less the distance, 
# similar the movies will be and hence can be recommended to similar users
%env JOBLIB_TEMP_FOLDER=/tmp

# define model
model_knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
# fit
model_knn.fit(movie_user_mat_sparse)
```

```python
# we will define two functions to perform the recommendation
# fuzzy_match to match the movie title over a fuzzy search
# perform_recommendation to perform the identify the movies to be recommended
def fuzzy_match(mapper, fav_movie, verbose=True):
    
    match_tuple = []
    # get match
    for title, idx in mapper.items():
        ratio = fuzz.ratio(title.lower(), fav_movie.lower())
        if ratio >= 60:
            match_tuple.append((title, idx, ratio))
    # sort
    match_tuple = sorted(match_tuple, key=lambda x: x[2])[::-1]
    if not match_tuple:
        print('Oops! No match is found')
        return
    return match_tuple[0][1]

def perform_recommendation(model_knn, data, mapper, fav_movie, no_of_recomendation):
    
    # fit
    model_knn.fit(data)
    # get input movie index
    print('Input Movie Name :', fav_movie, '\n')
    idx = fuzzy_match(mapper, fav_movie, verbose=True)
    # inference
    print('Recommendation system start to make inference')
    print('......\n')
    distances, indices = model_knn.kneighbors(data[idx], n_neighbors=no_of_recomendation+1)
    # get list of raw idx of recommendations
    raw_recommends = \
        sorted(list(zip(indices.squeeze().tolist(), distances.squeeze().tolist())), key=lambda x: x[1])[:0:-1]
    # get reverse mapper
    reverse_mapper = {v: k for k, v in mapper.items()}
    # print recommendations
    print('Recommendations for {}:'.format(fav_movie))
    for i, (idx, dist) in enumerate(raw_recommends):
        print('{0}: {1}, with distance of {2}'.format(i+1, reverse_mapper[idx], dist))
```

```python
# let's test our rec system for Movie16 that has been rated few times and identify
# similar movies that can be recommended.
my_favorite = 'Movie16'

perform_recommendation(
    model_knn=model_knn,
    data=movie_user_mat_sparse,
    fav_movie=my_favorite,
    mapper=movie_to_idx,
    no_of_recomendation=5)
```

#### <font color=green> We can see that the rec system is able to identify similar movies and recommend accordingly.
