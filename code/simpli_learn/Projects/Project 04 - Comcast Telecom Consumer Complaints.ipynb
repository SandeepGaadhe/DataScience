#### <font color=red> Customer Complaint are entered in different case sensitivity resulting in incorrect counts (eg. "Comcast" & "comcast")

```python
# using fuzzywuzzy to group similar items correctly
distinct_customer_complaint = df_rawdata['Customer Complaint'].unique()

# fuzz is used to compare TWO strings
from fuzzywuzzy import fuzz

# process is used to compare a string to MULTIPLE other strings
from fuzzywuzzy import process

# see sample of fuzzy match
print(process.extract("Comcast", distinct_customer_complaint, limit=30, scorer=fuzz.token_sort_ratio))
```

```python
# see sample of fuzzy match for another customer complaint
print(process.extract("Comcast Data Caps", distinct_customer_complaint, limit=30, scorer=fuzz.token_sort_ratio))
```

#### <font color=orange> Grouping customer complaint on fuzzy match

```python
# we will add a duplicate column for "Customer Complaint" as "Customer Complaint Type" 
# and will update/categorize the values in new columns based on fuzzy match
df_rawdata['ComplaintType'] = df_rawdata['Customer Complaint']
df_rawdata.head()
```

```python
# categorizing complaints in column ComplaintType based on fuzzy match in column "Customer Complaint"
import time

fuzzy_matchthreshold = 75

curr_iteration = 0
total_iteration = distinct_customer_complaint.size

for customer_complaint in distinct_customer_complaint:
    
    # print timer to know things are working
    start_time = time.process_time()
    curr_iteration += 1
    
    # fuzzy sort search
    cust_comp_sort_variants = process.extract(customer_complaint, distinct_customer_complaint, limit=5, scorer=fuzz.token_sort_ratio)
    # fuzzy set search
    cust_comp_set_variants = process.extract(customer_complaint, distinct_customer_complaint, limit=5, scorer=fuzz.token_set_ratio)

    # filter out fuzzy search below threshold value
    cust_comp_sort_var_selected = [cust_comp_sort_variant for cust_comp_sort_variant in cust_comp_sort_variants \
                                  if cust_comp_sort_variant[1] >= fuzzy_matchthreshold]
    cust_comp_set_var_selected = [cust_comp_set_variant for cust_comp_set_variant in cust_comp_set_variants \
                                  if cust_comp_set_variant[1] >= fuzzy_matchthreshold]

    # combine the filtered cust complaint variants
    cust_comp_var_selected = cust_comp_sort_var_selected + cust_comp_set_var_selected
    
    # use the first value to assign category
    comptype_toBeAssigned = cust_comp_var_selected[0][0]
    
    for cust_comp_var in cust_comp_var_selected:
        #df.loc[selection criteria, columns I want] = value
        df_rawdata.loc[df_rawdata.ComplaintType == cust_comp_var[0], 'ComplaintType'] = comptype_toBeAssigned
    
    if (curr_iteration % 100 == 0):
        elapsed_time = time.process_time() - start_time
        print('elapsed_time for batch of', curr_iteration ,': ', elapsed_time)

    
```

## <font color=blue>Analysis Tasks - 02

Which complaint types are maximum i.e., around internet, network issues, or across any other domains.
- Create a new categorical variable with value as Open and Closed. Open & Pending is to be categorized as Open and Closed & Solved is to be categorized as Closed.
- Provide state wise status of complaints in a stacked bar chart. Use the categorized variable from Q3. Provide insights on:

- Which state has the maximum complaints
- Which state has the highest percentage of unresolved complaints

- Provide the percentage of complaints resolved till date, which were received through the Internet and customer care calls.

```python
# new categorical variable with value as Open and Closed
df_rawdata["newStatus"] = ["Open" if Status=="Open" or Status=="Pending" else "Closed" for Status in df_rawdata["Status"]]
```

```python
# group results by new status column
df_rawdata["newStatus"].value_counts()
```

```python
# state with maximum complaints stacked bar chart
Status_complaints = df_rawdata.groupby(["State","newStatus"]).size().unstack().fillna(0)
Status_complaints.plot(kind="barh", figsize=(30,50), stacked=True)
```

```python
# state with maximum complaints
df_rawdata.groupby(["State"]).size().sort_values(ascending=False).to_frame().reset_index().rename({0: "Count"}, axis=1)[:5]
```

```python
# state with highest percentage of unresolved complaints
Status_complaints['UnresolvedPercent'] = ((Status_complaints['Open'].astype(int))/(Status_complaints['Closed'].astype(int))*100)

# we will apply a threahold of minimum 10 open requests to apply sorting
Status_complaints.query('Open >= 10').sort_values(by = 'UnresolvedPercent', ascending = False)
```

```python
#  percentage of complaints resolved till date, which were received through the Internet and customer care calls
#df_rawdata.groupby(["Received Via","newStatus"]).size()

total_calls = df_rawdata[(df_rawdata['Received Via'] == 'Customer Care Call')].shape[0]
resolved_calls = df_rawdata[(df_rawdata['Received Via'] == 'Customer Care Call') & (df_rawdata['newStatus'] == 'Closed') ].shape[0]
percent_resolved_calls = (resolved_calls/total_calls)*100
print('total calls RECEIVED via Customer Care Call : ', total_calls)
print('total calls RESOLVED via Customer Care Call : ', resolved_calls)
print('perct calls RESOLVED via Customer Care Call : ', percent_resolved_calls)

total_calls = df_rawdata[(df_rawdata['Received Via'] == 'Internet')].shape[0]
resolved_calls = df_rawdata[(df_rawdata['Received Via'] == 'Internet') & (df_rawdata['newStatus'] == 'Closed') ].shape[0]
percent_resolved_calls = (resolved_calls/total_calls)*100
print('\n')
print('total calls RECEIVED via Internet : ', total_calls)
print('total calls RESOLVED via Internet : ', resolved_calls)
print('perct calls RESOLVED via Internet : ', percent_resolved_calls)
```
